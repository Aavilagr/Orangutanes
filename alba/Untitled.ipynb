```python
#Removing "observaciones"
df1=df_raw.drop(labels="Observaciones", axis=1)
df1
```

```python
# Importing LabelEncoder from Sklearn
# library from preprocessing Module.
from sklearn.preprocessing import LabelEncoder
 
# Creating a instance of label Encoder.
le = LabelEncoder()
 
# Using .fit_transform function to fit label
# encoder and return encoded label
label = le.fit_transform(df1['Sujeto'])
 
# printing label
label
```

```python
## Appending the array to our dataFrame
# with column name 'Sujeto_number'
df1["Sujeto_number"] = label
```

```python
#Checking number each subject
df1.loc[df1['Sujeto'] == "Popo"] #4
df1.loc[df1['Sujeto'] == "Hadia"] #0
df1.loc[df1['Sujeto'] == "Storma"] #5
df1.loc[df1['Sujeto'] == "Jawie"] #1
df1.loc[df1['Sujeto'] == "Locky"] #3
df1.loc[df1['Sujeto'] == "Karl"] #2
```

```python
# Drop sujeto and ubicacion column with names

df2=df1.drop(labels=["Sujeto","Ubicacion"], axis=1)
```

```python
#Convert Nan in frecuencia to 0.
df2['Frecuencia'] = df2['Frecuencia'].fillna(0)
```
